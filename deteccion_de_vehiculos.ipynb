# Algoritmo de detección de vehiculos

```python
import torch
print(torch.__version__)
import cv2
from ultralytics import YOLO
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(f"Usando dispositivo: {device}")
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5l', pretrained=True).to(device)
model.conf = 0.4   
if device.type == 'cuda':
    model.half()   

cap = cv2.VideoCapture("deteccion_de_vehiculos.mp4")
if not cap.isOpened():
    print("Error al abrir el video")
    exit(1)

# Obtener resolución original y definir tamaño de visualización
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(f"Resolución original: {width}x{height}")
disp_w = 640
disp_h = int(disp_w * height / width)

total_frames = 0
vehicle_count = 0
person_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    total_frames += 1

    # Preprocesado
    frame_enh = cv2.convertScaleAbs(frame, alpha=1.3, beta=30)

    # Inferencia
    img = cv2.cvtColor(frame_enh, cv2.COLOR_BGR2RGB)
    results = model(img, size=960)

    # Procesar detecciones
    for *box, conf, cls in results.xyxy[0]:
        x1, y1, x2, y2 = map(int, box)
        label = model.names[int(cls)]

        if label in ['car', 'truck', 'bus', 'motorbike']:
            vehicle_count += 1
            color = (255, 0, 0)
        elif label == 'person':
            person_count += 1
            color = (0, 255, 0)
        else:
            continue

        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Redimensionar para mostrarse a 640px de ancho
    frame_disp = cv2.resize(frame, (disp_w, disp_h))
    cv2.imshow("Detección de Vehículos y Personas", frame_disp)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()

print(f"Total de frames procesados: {total_frames}")
print(f"Personas detectadas: {person_count}")
print(f"Vehículos detectados: {vehicle_count}")

```
